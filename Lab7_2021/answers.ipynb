## Ζήτημα 1.

Για κάθε $x\in(-1,1)$ έχουμε ότι:
$$
\sum_{k=0}^{\infty} x^k = \frac{1}{1-x} \to  (\sum_{k=0}^{\infty} x^k)' = \sum_{k=1}^{\infty}  kx^{k-1} = \frac{1}{(1-x)^2}
$$

Πολλαπλασιάζοντας και τα δύο μέλη με $x$ παίρνουμε:
$$
 \sum_{k=1}^{\infty}  kx^{k} =  \sum_{k=0}^{\infty}  kx^{k} =\frac{x}{(1-x)^2}
$$

Θέτουμε $x=1/2$:
$$
  \sum_{k=0}^{\infty} \frac{k}{2^k} = 2
$$

Τέλος πολλαπλασιάζουμε και τα 2 μέλη με $1/2$ και παίρνουμε το ζητούμενο:
$$
  \sum_{k=0}^{\infty} \frac{k}{2^{k+1}} = 1 = \sum_{k=0}^{\infty} k \pi(k)
$$


```python
import numpy as np


def random_walk_next(state):
    if np.random.uniform() < 0.5:
        return 0
    return state + 1

N = 10**4
running_state = 0
sum_result = 0

for i in range(N):
    running_state = random_walk_next(running_state)
    sum_result += running_state

    
### Ergodic Limit Theorem
est = sum_result/N
print("The simulated ergodic average [X1+X2+X3+...+XN]/N is: %.4f" % (est))
act = 1.0
err = 100*abs(1-est/act)
print("The relative error is {:.4f}%".format(err))
```

```python
import numpy as np
import statistics as stat


def random_walk_next(state):
    if np.random.uniform() < 0.5:
        return 0
    return state + 1

N = 10**4
s = []

for _ in range(50):
    
    running_state = 0
    sum_result = 0

    for i in range(N):
        running_state = random_walk_next(running_state)
        sum_result += running_state
    s.append(sum_result)

s = [i/N for i in s]
### Ergodic Limit Theorem
print("Simulated ergodic average\nMean : {:.4f}\nVariance : {:.4f}".format(stat.mean(s),stat.
                                                                           variance(s))
     )

print("Standard deviation: {:.4f}".format(stat.stdev(s)))
```

## Ζήτημα 3.
Για να μειωθεί η δειγματική τυπική απόκλιση στο μισό, θα πρέπει η δειγματική διασπορά να υποτετραπλασιαστεί, δηλαδή το δείγμα θα πρέπει να διπλασιαστεί.

```python
import numpy as np
import statistics as stat


def random_walk_next(state):
    if np.random.uniform() < 0.5:
        return 0
    return state + 1

N = 2*10**4
s = []

for _ in range(50):
    
    running_state = 0
    sum_result = 0

    for i in range(N):
        running_state = random_walk_next(running_state)
        sum_result += running_state
    s.append(sum_result)

s = [i/N for i in s]
### Ergodic Limit Theorem
print("Simulated ergodic average\nMean : {:.4f}\nVariance : {:.4f}".format(stat.mean(s),stat.
                                                                           variance(s))
     )
print("Standard deviation: {:.4f}".format(stat.stdev(s)))
```

## Ζήτημα 4.

Έστω $f:X\to R$ με τύπο $f(x)=cos(x+cos(x))$. Η f είναι φραγμένη επί του $[-1,1]$, επομένως ισχύουν οι προϋποθέσεις του Θ. 31 και για την αλυσίδα του Ζητήματος 1 έχουμε ότι:
$$
\frac{f(X_1)+f(X_2)+...+f(X_N)}{N} \to \sum_{k=0}^{\infty} f(k)*π(k) = \sum_{k=0}^{\infty} \frac{cos(k+cos(k))}{2^{k+1}}
$$

Αρκεί λοιπόν να πολλαπλασιάσουμε την παραπάνω ποσότητα με $2$ ώστε να πάρουμε το ζητούμενο άθροισμα.

```python
import numpy as np
from math import cos


def random_walk_next(state):
    if np.random.uniform() < 0.5:
        return 0
    return state + 1

N = 10**4
running_state = 0
I = cos(1)
sum_result = random_walk_next(running_state)

for i in range(1,N):
    running_state = random_walk_next(running_state)
    sum_result += cos(running_state+cos(running_state))
    
### Ergodic Limit Theorem
est = sum_result/N * 2
print("Η εκτίμηση του αθροίσματος είναι: %.4f" % (est))
act = 0.0

for i in range(1000):
    act += cos(i+cos(i))/(2**(i))
print("Το άθροισμα των 1000 πρώτων όρων είναι: {:.4f}".format(act))

err = 100*abs(1-est/act)
print("Το σχετικό σφάλμα είναι: {:.4f}%".format(err))
```

# Παραδοτέο 2

Καθώς στο προηγούμενο εργαστήριο είχα αυξήσει των αριθμό των δειγμάτων σε $10000$, για να είναι δίκαιη η σύγκριση θα πρέπει η αλυσίδα να κάνει $Μ=n*N=10,000*100=10^6$ βήματα στον κύλινδρο $C_d$. Τα σφάλματα με αυτή την μέθοδο είναι πολύ μικρότερα και οι εκτιμήσεις μας είναι πολύ πιο κοντά στις θεωρητικά αναμενόμενα τιμές, σε σχέση με αυτές του προηγούμενου εργαστηρίου.

```python
from math import pi, gamma
from numpy import random as r
from numpy import array

def Vol1(d):
    x = d/2
    return pi ** x / gamma(x + 1)
######################################

delta = 1.0
N = 100 # number of steps to approach equilibrium
samples = 10000
V_old = 2 # 1-D ball Volume
#######################################
M = N*samples

for d in range(1,100):
    Nhits = 0
    x = [0,]*d  ## start at the centre of th disc. This variable will keep the position of the chain
    R_sq = 0.0  ## this variable keeps the squared distance from 0. It saves some computations to keep it
    for i in range(M):
        k = r.choice(range(d))  ## choose a jump direction at random
        z = r.uniform(-delta,delta) ## choose a jump size uniformly in (-delta,delta)
        x_prop_k = x[k] + z   ## propose a jump by z in the direction k
        R_sqprop = R_sq - x[k]**2+ x_prop_k**2 ## compute the squared distance from 0 after the proposed jump 
        if R_sqprop < 1.0: 
            R_sq = R_sqprop
            x[k]= x_prop_k   ## if the proposed jump leads to a point in the disc, then jump
        
        y = [*x, r.uniform(-1,1)] ## Choose a new component randomly between[-1,1]
        if sum(array(y)**2) <1: Nhits +=1 ## Check if the new point is inside the d+1 dimension-unitary sphere 
            
    V_new = Nhits/M * 2 * V_old
    V_old = V_new
    V_act = Vol1(d+1)
    err = 100*abs(1-V_new/V_act)
    print("{}-dimensions unitary sphere volume\nEst.: {:.5}\nAct.: {:.5}\nErr.: {:.5}%\n".format(d+1,V_new,V_act,err))
```
